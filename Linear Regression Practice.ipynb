```python
import pandas as pd
df = pd.read_csv("D:\ML_data_set\Salary Data.csv")
df
```

```python
x,y = df[["Experience Years"]], df[["Salary"]]
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
x.mean()
```

```python
y.mean()
```

```python
from matplotlib import pyplot as plt
plt.scatter(x.mean(),y.mean(), color = "red")
plt.scatter(df["Experience Years"], df["Salary"])
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.title("Linear_Regression_data_set")
```

```python
reg.fit(x,y)
```

```python
reg.predict([[10.4]])
```

```python
reg.score(x,y)
```

```python
df["predicted_y"] = reg.predict(x)
df
```

```python
plt.plot(x,reg.predict(x))
plt.scatter(x.mean(),y.mean(), color = "red")
plt.scatter(df["Experience Years"], df["Salary"])
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.title("Linear_Regression_data_set")
```
