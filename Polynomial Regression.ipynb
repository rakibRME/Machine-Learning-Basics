```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Import Linear dataset

```python
df1 = pd.read_csv("D:\ML_data_set\Linear Regression - Sheet1.csv")
df1.head()
```

```python
x_linear = df1[['X']]
y_linear = df1[['Y']]
```

# Import Non-linear dataset

```python
df2 = pd.read_csv("D:\ML_data_set\Ice_cream selling data.csv")
df2.head()
```

```python
x_nonlinear = df2[['Temperature (°C)']]
y_nonlinear = df2[['Ice Cream Sales (units)']]
```

# Visual Representation

```python
#linear
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(x_linear, y_linear, color='blue', label='Data with Linear Relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Relationship')
plt.legend()

#non-linear
plt.subplot(1, 2, 2)
plt.scatter(x_nonlinear, y_nonlinear, color='green', label='Data with Non-linear Relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Non-linear Relationship')
plt.legend()

plt.tight_layout()
plt.show()
```

# Linear

```python
from sklearn.linear_model import LinearRegression
reg1 = LinearRegression()
```

```python
reg1.fit(x_linear, y_linear) # x= x_linear.reshape(-1, 1)
```

```python
reg1.score(x_linear , y_linear)
```

```python
plt.scatter(x_linear, y_linear, color='blue', label='Data with Linear Relationship')
plt.plot(x_linear, reg1.predict(x_linear), color='red', label='Best fit line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Relationship')
plt.legend()
```

# Non Linear

```python
reg2 = LinearRegression()
reg2.fit(x_nonlinear , y_nonlinear)
```

```python
reg2.score(x_nonlinear, y_nonlinear)
```

```python

plt.scatter(x_nonlinear, y_nonlinear, color='green', label='Data with Non-linear Relationship')
plt.plot(x_nonlinear, reg2.predict(x_nonlinear), color='red', label='Best fit line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Non-linear Relationship')
plt.legend()
```
