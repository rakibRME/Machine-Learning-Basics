```python
age_stats = df.groupby(['race', 'sex'])['age'].describe()
print(age_stats)

max_age_amer_indian_eskimo_men = df[(df['race'] == 'Amer-Indian-Eskimo') & (df['sex'] == 'Male')]['age'].max()
print("Maximum age of Amer-Indian-Eskimo men:", max_age_amer_indian_eskimo_men)
```

```python
men = df[df['sex'] == 'Male']
men['marital_status_group'] = men['marital-status'].apply(lambda x: 'married' if x.startswith('Married') else 'single')
proportion = men.groupby('marital_status_group')['salary'].apply(lambda x: (x == '>50K').mean())
print(proportion)
```

```python
max_hours_per_week = df['hours-per-week'].max()
print("Maximum hours worked per week:", max_hours_per_week)
people_working_max_hours = df[df['hours-per-week'] == max_hours_per_week]
num_people_working_max_hours = len(people_working_max_hours)
print("Number of people working", max_hours_per_week, "hours per week:", num_people_working_max_hours)
percentage_earning_more_than_50K = (people_working_max_hours['salary'] == '>50K').mean() * 100
print(f"Percentage of people earning >50K among those working {max_hours_per_week} hours: {percentage_earning_more_than_50K:.2f}%")
```

```python
avg_hours_by_country_salary = df.groupby(['native-country', 'salary'])['hours-per-week'].mean().reset_index()
avg_hours_japan = avg_hours_by_country_salary[avg_hours_by_country_salary['native-country'] == 'Japan']
print(avg_hours_japan)
```
