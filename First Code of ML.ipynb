```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("D:\ML_data_set\Linear_Regression_data_set.csv")
df
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns = ['Date'])
df
```

```python
plt.scatter(df["Starting(USD)"], df["Ending(USD)"])
plt.xlabel("Starting(USD)")
plt.ylabel("Ending(USD)")
plt.title("Linear_Regression_data_set")
```

```python
x = df.drop(columns = ["Ending(USD)"])
x
```

```python
y = df[["Ending(USD)"]]
y
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
x.mean()
```

```python
y.mean()
```

```python
plt.scatter(x.mean(),y.mean(), color = "red")
plt.scatter(df["Starting(USD)"], df["Ending(USD)"])
plt.xlabel("Starting(USD)")
plt.ylabel("Ending(USD)")
plt.title("Linear_Regression_data_set")
```

```python
reg.fit(x,y)
```

```python
m = reg.coef_
m
```

```python
c = reg.intercept_
c
```

```python
m*16700 + c
```

```python
reg.predict([[16700]])
```

```python
df["predicted_y"] = reg.predict(x)
df
```

```python
plt.plot(x,reg.predict(x))
plt.scatter(x.mean(),y.mean(), color = "red")
plt.scatter(df["Starting(USD)"], df["Ending(USD)"])
plt.xlabel("Starting(USD)")
plt.ylabel("Ending(USD)")
plt.title("Linear_Regression_data_set")
```

```python
reg.predict([[16600]])
```

```python
df
```

```python
lose = df["Ending(USD)"] - df["predicted_y"]
df["Lose"] = lose
df
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mse = mean_squared_error(df["Ending(USD)"], df["predicted_y"])
mse
```

```python
mae = mean_absolute_error(df["Ending(USD)"], df["predicted_y"])
mae
```

```python
mse = abs(df["Lose"])
mse
#sum(mse)/len(x)
```

```python
reg.score(x,y)
```
