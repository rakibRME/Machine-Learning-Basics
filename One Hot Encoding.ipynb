```python
import pandas as pd
```

```python
data = {"category" : ["A", "B", "C","A", "B", "C","A", "B", "C"]}
```

```python
df = pd.DataFrame(data)
df
```

```python
on_hot_encoded_df = pd.get_dummies(df, columns=["category"])
on_hot_encoded_df
```

```python
on_hot_encoded_df = pd.get_dummies(df, columns=["category"], prefix="Column",)
on_hot_encoded_df
```

```python
on_hot_encoded_df = pd.get_dummies(df, columns=["category"], prefix="Column",drop_first= True,)
on_hot_encoded_df
```
